```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
from statsmodels.tsa.statespace.sarimax import SARIMAX
import os
import csv


def mean_absolute_percentage_error(y_true, y_pred): 
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

def extract_lat_lon(filename):
    # Regular expression pattern to match latitude and longitude in the filename
    pattern = re.compile(r'(-?\d+(\.\d+)?)')
    
    # Search for the pattern in the filename
    matches = pattern.findall(filename)
    
    # If the pattern is found, extract latitude and longitude
    if matches:
        latitude = float(matches[0][0])
        longitude = float(matches[1][0])
        return latitude, longitude
    else:
        return None, None


folder_path = 'SeaData\\csv\\d10all'
date = 'd10'

with open(f"Statistics for SARIMAX {date}.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow([f'Statistics for SARIMAX {date}'])
    
    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        print(f'Processing file: {file_path}')
        tag = 'dbiastg' if 'dbiastg' in file_name else 'biastg'
        writer.writerow([file_name])       

        data = pd.read_csv(file_path, index_col='time', parse_dates=True)

        # Prepare the data
        data = data[tag].asfreq('B').fillna(method='ffill')  # Fill missing values

        # Split the data into training and testing sets
        train_size = int(len(data) * 0.8)
        train_data, test_data = data[:train_size], data[train_size:]

        # Create and fit the SARIMAX model
        model = SARIMAX(train_data, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
        results = model.fit(disp=-1)
        
        # Make predictions
        test_predict = results.forecast(steps=len(test_data))
        
        latitude, longitude = extract_lat_lon(file_name)

        # Evaluate the model
        mae = mean_absolute_error(test_data, test_predict)
        mse = mean_squared_error(test_data, test_predict)
        rmse = np.sqrt(mse)
        mbe = np.mean(test_data - test_predict)
        mape = mean_absolute_percentage_error(test_data, test_predict)
        
        writer.writerow(['latitude', latitude])
        writer.writerow(['longitude', longitude])
        print(f"RMSE: {rmse:.2f}")
        writer.writerow(['RMSE', rmse])
        print(f"MAE: {mae:.2f}")
        writer.writerow(['MAE', mae])
        print(f"MBE: {mbe:.2f}")
        writer.writerow(['MBE', mbe])
        print(f"MAPE: {mape:.2f}%")
        writer.writerow(['MAPE', mape])

        # Plot the results
#         plt.figure(figsize=(14, 6))
#         plt.plot(data, label='Observed')
#         plt.plot(test_data.index, test_predict, color='r', label='Predicted')
#         plt.legend()
#         plt.show()
```
