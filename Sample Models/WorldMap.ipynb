```python
import plotly.express as px
import pandas as pd
import csv

def prepare_dataset(file_path,index):
    with open(file_path, 'r') as file:
        lines = csv.reader(file, delimiter='\t')
        lines = list(lines)
#         print(lines)

    data = []
    for i in range(len(lines)):
        line = lines[i]
        if len(line) > 0 and line[0].endswith('.csv'):
            latitude = float(lines[i+1][0].split(',')[1])
            longitude = float(lines[i+2][0].split(',')[1])
            RMSE = float(lines[i+index+3][0].split(',')[1])
            data.append((latitude, longitude, RMSE))
    df = pd.DataFrame(data, columns=['latitude', 'longitude', 'RMSE'])
    return df

Markers = ['RMSE','MAE','MBE','MAPE']
for index, PMarker in enumerate(Markers):
    Data = prepare_dataset('Results 7-03//Statistics for BiDirectional LSTM 04-07-2023.csv',index)
    Data[PMarker]=Data['RMSE']
    # print(Data)
    # Get the minimum and maximum RMSE values
    min_rmse = Data['RMSE'].min()
    max_rmse = Data['RMSE'].max()
    title=f'BiDirectional LSTM-TimeSeries Prediction Model Performance--{PMarker}'
    # Create a scatter geo plot
    fig = px.scatter_geo(Data,
                         lat='latitude',
                         lon='longitude',
                         color=PMarker,  # Use 'RMSE' column for color
                         color_continuous_scale=px.colors.sequential.Plasma,
                         range_color=[min_rmse, max_rmse],
                         template='plotly',
                         title=title,
                         width=1000,  # Set the width of the plot
                         height=800,  # Set the height of the plot
                         size_max=0.6)  # Set the maximum marker size

    # Save the image as a PNG file
    fig.write_image(f"Results 7-03//{title}.png")

    # Display the plot in a web browser
    fig.show()
```

```python
import pandas as pd
import re

def update_lat_lon(file_path):
    # Read the file line by line
    with open(file_path, 'r') as file:
        lines = file.readlines()

    # Iterate through each line in the file
    for i, line in enumerate(lines):
        # Check if the line contains a file name
        if '.csv' in line:
            # Extract latitude and longitude from the file name
            lat, lon = re.findall(r'lat_(-?\d+(\.\d+)?)_lon_(-?\d+(\.\d+)?)', line)[0][:3:2]
            # Update the latitude and longitude in the lines
            lines[i + 1] = f'latitude,{lat}\n'
            lines[i + 2] = f'longitude,{lon}\n'

    # Write the updated lines back to the file
    with open(file_path, 'w') as file:
        file.writelines(lines)

# Specify the path to your CSV file
file_path = 'Results//StatisticsforSARIMAXd10.csv'

# Call the function with the file path
update_lat_lon(file_path)
```
